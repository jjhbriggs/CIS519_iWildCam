```python
import codecs
```

```python
with codecs.open("input/metadata/metadata/iwildcam2022_mdv4_detections.json", 'r',
                 encoding='utf-8', errors='ignore') as f:
    md_meta = json.load(f)
```

```python
# simplify annotation frame
df_label = pd.read_csv("input/metadata/metadata/train_sequence_counts.csv")

f = open('input/metadata/metadata/iwildcam2022_train_annotations.json')
with codecs.open("input/metadata/metadata/iwildcam2022_train_annotations.json", 'r',
                 encoding='utf-8', errors='ignore') as f:
    train_meta = json.load(f)
df_seq = pd.DataFrame(train_meta['images'])
df_seq = df_seq.drop(["seq_num_frames","datetime","width","width","height","id","sub_location","location"],axis=1)
df_seq['count_per_image'] = [0]*len(df_seq)
df_seq = df_seq.where(df_seq['seq_id'].isin(list(df_label['seq_id']))).dropna()
df_seq = df_seq.reset_index(drop=True)
df_seq.head()
```

```python
from tqdm import tqdm
for i in tqdm(range(len(df_label))):
  seq_id = df_label.iloc[i]['seq_id']
  count = df_label.iloc[i]['count']
  df_seq['count_per_image'] = np.where(df_seq['seq_id']==seq_id, count, df_seq['count_per_image'])
```

```python
df_seq["file_name"] = 'train/' + df_seq['file_name'].astype(str)
```

```python
df_mega_empty = pd.DataFrame(md_meta['images'])
df_mega_empty = df_mega_empty.where(~df_mega_empty['file'].isin(list(df_seq['file_name']))).dropna()
df_mega_empty_train = df_mega_empty[df_mega_empty['file'].str.startswith('train', na=False)]
df_mega_empty_train = df_mega_empty_train[df_mega_empty_train['detections'].map(len) == 0]
```

```python
df_mega_empty_train_5000 = df_mega_empty_train.sample(5000)
df_mega_empty_train_5000 = df_mega_empty_train_5000.reset_index(drop=True)
df_mega_empty_train_5000
```

```python
for index, row in df_mega_empty_train_5000.iterrows():
  if index % 500 == 0:
    print("moved:",index)
  ! cp {os.path.join('input/train/',row['file'])} {os.path.join('empty_augment/',row['file'])}
```
